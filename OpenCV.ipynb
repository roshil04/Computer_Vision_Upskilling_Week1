# OpenCV


## Introduction


[OpenCV](https://opencv.org/)(Open Source Computer Vision Library) is an open-source computer vision and machine learning library, under the BSD license developed by Intel in 1996. It also has features of cross-platform, i.e., you can use it on multiple computing platforms(environment in which a piece of software is executed either may be operating system(OS) like Windows, Linux, or, hardware like Wintel, Macintosh). 


## Features and possible operations 


It has a modular structure,i.e., it's packages include several shared or static libraries.  The modules available in the CV are as follows:

* [Core functionality](https://docs.opencv.org/4.3.0/d0/de1/group__core.html)(core): 

  It is a compact module defining basic data structure and dense multi-dimensional array and basic functions used by all other modules.

* [Image Processing](https://docs.opencv.org/4.3.0/d7/dbd/group__imgproc.html)(imgproc):

  It is an image processing module that includes linear and non-linear image filtering, geometrical image transformation, color space conversion, histograms, etc.

* [Video Analysis](https://docs.opencv.org/4.3.0/d7/de9/group__video.html)(video): 

  It is a video analysis module that includes motion estimation, background subtraction, and object tracking algorithms.

* [Camera Calibration and 3D Reconstruction](https://docs.opencv.org/4.3.0/d9/d0c/group__calib3d.html)(calib3d):

  It includes basic multiple-view geometry algorithms, single and stereo camera calibration, object pose estimation, stereo correspondence algorithms, and 3D reconstruction elements.

* [2D Features Framework](https://docs.opencv.org/4.3.0/da/d9b/group__features2d.html)(features2d):

  It includes salient feature detectors, descriptors, and descriptor matchers.

* [Object Detection](https://docs.opencv.org/4.3.0/d5/d54/group__objdetect.html)(objdetect): 

  It includes the detection of objects and instances of the predefined classes (for example, faces, eyes, mugs, people, cars, and so on).


* [High-level GUI](https://docs.opencv.org/4.3.0/d7/dfc/group__highgui.html)(highgui) 

 It is an easy-to-use interface to simple UI capabilities.

* [Video I/O](https://docs.opencv.org/4.3.0/dd/de7/group__videoio.html)(videoio):

  It is an easy-to-use interface for video capturing and video codecs.

There are other helper modules, such as `FLANN` and `Google test wrappers`, Python bindings, and others.

## Image I/O 


```python
# Importing OpenCV and others Libraries
import pandas as pd
import numpy as np
import cv2
import matplotlib.pyplot as plt
from sklearn.decomposition import FastICA
from IPython.display import display, Image
```

```python
## loading image
img_path = 'kiki.jpg'

# Reading the image using OpenCV
image = cv2.imread(img_path) ## loads image from the passed path, img_path
```

```python
# Checking if the image was read successfully
if image is None:
    print("Image not found or invalid format")
else:
    # Display the image using IPython.display
    display(Image(filename='kiki.jpg'))
```

```python
## displaying the image
plt.imshow(image)
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
plt.close()
```

Here, you can see that `matplotlib.pyplot` is used for displaying the images instead of `cv2.imshow()`. Both can be used to view the images. The only difference is in how opencv handles images. The default way opencv loads image is in the BGR format (Blue, Green and Red), while `matplotlib` loads images in the RGB format (Red, Green and Blue). Due to this, the image shown above has it's blue and green channel reversed.


```python
#To view the actual image
# converting the BGR color space to RGB colorspace
bgr_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(bgr_image)
plt.xticks([])
plt.yticks([])
plt.show()
plt.close()
```

```python
#Resizing and rescaling the image

# Load an image from file
image = cv2.imread('kiki.jpg')

# Defining the new width and height dimensions
new_width = 400
new_height = 300

# Choose the interpolation method
interpolation = cv2.INTER_LINEAR  # We can also choose cv2.INTER_NEAREST, cv2.INTER_LINEAR, cv2.INTER_CUBIC, etc.

# Resize the image with the specified dimensions and interpolation method
resized_image = cv2.resize(image, (new_width, new_height), interpolation=interpolation)

# Display the resized image using Matplotlib
plt.imshow(cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
```

```python
# Rotating the image

# Load an image from file
image = cv2.imread('kiki.jpg')

# Defining the angle by which we want to rotate the image in degrees
angle = 45  

# Geting the center of the image 
height, width = image.shape[:2]
center = (width // 2, height // 2)

# Defining the rotation matrix
rotation_matrix = cv2.getRotationMatrix2D(center, angle, scale=1.0)

# Applying the rotation to the image using warpAffine
rotated_image = cv2.warpAffine(image, rotation_matrix, (width, height))

# Display the rotated image
plt.imshow(cv2.cvtColor(rotated_image, cv2.COLOR_BGR2RGB))
plt.axis('off')  # Hiding the axis
plt.show()
```
